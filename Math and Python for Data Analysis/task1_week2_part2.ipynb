```python
import math
from matplotlib import pylab as plt
import scipy.linalg as lg
import numpy as np
%matplotlib inline
```

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
def linear_system(list_of_x):
    list_A = []
    list_b = []
    for x in list_of_x:
        list_A.append([x**ex for ex in range(len(list_of_x))])
        list_b.append(f(x))
    
    return list_A, list_b
```

```python
def solve(list_of_x):
    (A, b) = linear_system(list_of_x)
    
    return lg.solve(A, b)
```

```python
def new_polynomial(list_of_x, x):
    solving = solve(list_of_x)
    print(solving)
    
    answer = np.zeros(len(x))
    for i in range(len(solving)):
        answer += solving[i]*x**i
    
    return answer
```

```python
x = np.arange(0, 16, 0.1)
list_of_x = [
    [1, 15], [1, 8, 15], [1, 4, 10, 15]
]
```

```python
y = np.vectorize(f)
plt.plot(x, y(x))
plt.show()
```

```python
y1 = new_polynomial(list_of_x[0], x)
plt.plot(x, y1)
plt.show()
```

```python
y2 = new_polynomial(list_of_x[1], x)
plt.plot(x, y2)
plt.show()
```

```python
y3 = new_polynomial(list_of_x[2], x)
plt.plot(x, y3)
plt.show()
```
